## **QUESTION 6 --->** Time Series Plots

```python
# Set 'order_date' as the index
df1.set_index('order_date', inplace=True)

# Resample the data to get total sales for each date
sales_time_series = df1['sales'].resample('D').sum()

# Plot the time series
plt.figure(figsize=(12, 6))
plt.plot(sales_time_series.index, sales_time_series, marker='o', linestyle='-')
plt.title('Sales Over Time')
plt.xlabel('Date')
```

## **QUESTION 7 --->** Seasonal Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Perform seasonal decomposition
decomposition = seasonal_decompose(sales_time_series, model='additive', period=365)  # Assuming yearly seasonality

# Plot the decomposed components
plt.figure(figsize=(12, 8))
plt.subplot(411)
plt.plot(decomposition.trend, label='Trend', color = 'purple')
plt.legend(loc='upper left')

plt.subplot(412)
plt.plot(decomposition.seasonal, label='Seasonality', color='salmon')
plt.legend(loc='upper left')

plt.subplot(413)
plt.plot(decomposition.resid, label='Residuals', color='gold')
plt.legend(loc='upper left')

plt.subplot(414)
plt.plot(sales_time_series, label='Original', color='skyblue')
plt.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

## **QUESTION 10 --->** Sales Variation Over Time

```python
# Resample the data to get total sales for each period
sales_monthly = df1['sales'].resample('M').sum()  # Monthly sales
sales_quarterly = df1['sales'].resample('Q').sum()  # Quarterly sales
sales_yearly = df1['sales'].resample('Y').sum()  # Yearly sales

# Plot line charts
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.plot(sales_monthly, marker='o', linestyle='-', color='b')
plt.title('Monthly Sales Variation')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(sales_quarterly, marker='o', linestyle='-', color='g')
plt.title('Quarterly Sales Variation')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(sales_yearly, marker='o', linestyle='-', color='r')
plt.title('Yearly Sales Variation')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## **QUESTION 10 --->** Discount Impact on Sales



â–¶ `*There is no 'discount' column in the dataset, I still analyze the impact of discounts on sales by comparing sales for different scenarios (e.g., with and without discounts). One way to do this is to create a new column that indicates whether a discount was applied to each transaction, and then compare the sales for transactions with and without discounts.*`

```python
# Create a new column 'discount_applied' to indicate whether a discount was applied
df1['discount_applied'] = df1['sales'].apply(lambda x: 'Yes' if x < 0 else 'No')

# Group the data by 'discount_applied' and sum the sales
sales_by_discount = df1.groupby('discount_applied')['sales'].sum()

# Plot bar chart
plt.figure(figsize=(8, 6))
sales_by_discount.plot(kind='bar')
```
